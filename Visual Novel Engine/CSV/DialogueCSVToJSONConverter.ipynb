```python
import os
import glob
import pandas as pd
import numpy as np
import json
```

```python
filekeys = open('dialoguekeys.txt').read().splitlines()
#filetextkeys = [x + '.txt' for x in filekeys]
print(filekeys)
```

```python
inputPath = 'E:/Users/William/Documents/GitHub/Visual Novel Engine/CSV'
outputPath = 'E:/Users/William/Documents/GitHub/Visual Novel Engine/Assets/RawData/DialogueData'
metaHeader = ['routeKey', 'setting']
dataHeader = ['id', 'route', 'speakerName', 'sprite', 'text', 'choices']

for key in filekeys:
    # Header
    header = pd.read_csv(key + '.csv', names=metaHeader, nrows=3)
    header = header.to_json(orient='records')
    #header = json.loads(header)
    print('\nHeader\n{0}'.format(header))
    
    # Dialogue
    dialogueFile = pd.read_csv(key + '.csv', skiprows=3, names=dataHeader, header=None)
    dialogueFile = dialogueFile.fillna('')
    dialogueFile = dialogueFile.to_dict(orient='records') # KEY CONVERSION
    dialogueDict = {"data": dialogueFile}
    #dialogueDict = json.loads(dialogueDict)
    
    print('\nFile Data\n{0}'.format(dialogueFile))
    
    # Print to file
    with open(outputPath + '/' + key + '_meta.json', "w+") as outfile:
        json.dump(header, outfile, indent=4)
        print('\nHeader Json\n{0}'.format(json.dumps(header, indent=4, separators=(',', ': '))))
        
    with open(outputPath + '/' + key + '_data.json', "w+") as outfile:
        json.dump(dialogueDict, outfile, indent=4)
        print('\nData Json\n{0}'.format(json.dumps(dialogueDict, indent=4, separators=(',', ': '))))
```
